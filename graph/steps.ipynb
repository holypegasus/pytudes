```python
# step-by-step visualize graph-exploration
import matplotlib.pyplot as plt
import networkx as nx

from oop import *


# visualizer
def highlight_explored(groups):
  # plot graph
  G = nx.Graph()
  for node0,node1 in tuple_graph:
    G.add_edge(node0, node1)
  # draw components
  pos = nx.spring_layout(G)

  nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
  nx.draw_networkx_edges(G, pos, edgelist=tuple_graph)
  nx.draw_networkx_labels(G, pos)
  # highlight explored nodes
  id_nodes_explored = set.union(*[group.nodes for group in groups])
  log(id_nodes_explored)
  nx.draw_networkx_nodes(G, pos, node_size=500, node_color='r', nodelist=id_nodes_explored)
  # plot
  plt.axis('off')

# test func
def explore_1(goal_groups):
  log('goal_groups: [%s]:\n%s', len(goal_groups), goal_groups)
  if terminal(goal_groups):
    log('Done exploring!')
  else:
    goal_groups = explore(goal_groups)
    highlight_explored(goal_groups)
  return goal_groups

# make graph
tuple_graph = [
  (1,2),
  (1,12),
  (2,3),
  (2,4),
  (3,9),
  (3,10),
  (4,5),
  (4,7),
  (4,10),
  (5,6),
  (5,7),
  (5,12),
  (6,11),
  (7,8),
  (7,13),
  (9,15),
  (11,14),
  ]
nodes = tuple_graph2nodes(tuple_graph)
goals = mapl(Node, (8,10,14))
goal_groups = [Group(goal) for goal in goals]
```
